```python
df = pd.read_csv('../data/diamonds_train.csv')
df
```

```python
# do some preprocessing!

df.cut.unique()
cut_encodeing = {
    "Fair" : 0,
    "Good" : 1,
    "Very Good" : 2,
    "Premium" : 3,
    "Ideal" : 4,
}

df['cut'].replace(cut_encodeing, inplace=True)
df
```

```python
df.color.unique()
color_encodeing = {
    "D" : 6,
    "E" : 5,
    "F" : 4,
    "G" : 3,
    "H" : 2,
    "I" : 1,
    "J" : 0,
}

df['color'].replace(color_encodeing, inplace=True)
df
```

```python
clarity_encodeing = {
    "I1" : 0,
    "SI2" : 1,
    "SI1" : 2,
    "VS2" : 3,
    "VS1" : 4,
    "VVS2" : 5,
    "VVS1" : 6,
    "IF" : 7,
}

df['clarity'].replace(clarity_encodeing, inplace=True)
df
```

```python
# modeling
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

train , test = train_test_split(df, test_size=.2, random_state=313)
train , valid = train_test_split(train, test_size=.2, random_state=313)

model = LinearRegression()
model.fit(train.drop('price', axis=1), train.price)
```

```python
print('intercept/bias = ', model.intercept_)
print('coefficients = ', model.coef_)
```

```python
# evaluate model
from sklearn.metrics import r2_score
y_pred = model.predict(valid.drop('price', axis=1))
r2_score(valid.price, y_pred)
```

# Modeling from scratch

$$X\beta = Y$$
$$((X^TX)^{-1}X^T)\ X\  \beta = ((X^TX)^{-1}X^T)\ Y$$
$$\beta = ((X^TX)^{-1}X^T)\ Y$$

```python
X , Y = train.drop('price', axis=1), train.price
X.insert(0, 'intercept',value=1)
X
```

```python
left_inverse = np.linalg.inv(X.T@X)@X.T
```

```python
beta = left_inverse @ Y
```

```python
print('intercept/bias = ', beta[0])
print('coefficients =\n', beta[1:])
```

```python
print('intercept/bias = ', model.intercept_)
print('coefficients = ', model.coef_)
```
