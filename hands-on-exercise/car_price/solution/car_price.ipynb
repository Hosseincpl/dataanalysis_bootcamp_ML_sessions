<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
<font face="vazir" size=3>
    در ابتدا نیاز است فایل‌های مجموعه‌داده را بخوانید. نمونه‌های آموزشی در فایل <code>car_train.csv</code> و نمونه‌های آزمون که باید دسته‌ی آن‌ها را پیش‌بینی کنید در فایل <code>car_test.csv</code> ذخیره شده‌اند. اگر لازم دانستید می‌توانید به دلخواه خود بخشی از مجموعه‌ی آموزشی را به عنوان مجموعه‌ی اعتبارسنجی نیز جدا کنید.

```python
df_train = pd.read_csv('data/car_train.csv')
df_train
```

```python
df_test = pd.read_csv('data/car_test.csv')
df_test
```

<h2 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
پیش‌پردازش و مهندسی ویژگی
</font>
</h2>

<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
<font face="vazir" size=3>
    در این سوال شما می‌توانید از هر تکنیک پیش‌پردازش/مهندسی ویژگی که در کلاس‌ها آموختید، استفاده کنید.
</font>
</p>

```python
df_train.isna().sum()
```

```python
df_test.isna().sum()
```

```python
df_train.duplicated().sum()
```

```python
df_train = df_train.drop_duplicates(keep = 'first')
df_train
```

```python
df_train[df_train['Number of Doors'].isna()]
```

```python
# For Ferrari FF 2013 there are 3 doors. check this link: https://en.wikipedia.org/wiki/Ferrari_FF#:~:text=It%20is%20Ferrari's%20first%20production,62%20mph
df_train.loc[(df_train['Number of Doors'].isna()) & (df_train['Make'] == 'Ferrari'), 'Number of Doors'] = 3

# For Tesla Model S there are 5 doors. check this link: http://en.wikipedia.org/wiki/Tesla_Model_S
df_train.loc[(df_train['Number of Doors'].isna())  & (df_train['Make'] == 'Tesla'), 'Number of Doors'] = 5
```

```python
df_test[df_test['Number of Doors'].isna()]
```

```python
df_test.loc[(df_test['Number of Doors'].isna())  & (df_test['Make'] == 'Tesla'), 'Number of Doors'] = 5
```

```python
df_train[df_train['Engine Fuel Type'].isna()]
```

```python
# Filled missing value of engine fuel type for Suzuki Verona as regular unleaded, as researched from this website: http://www.edmunds.com/suzuki/verona/2004/features-specs/
df_train.loc[(df_train['Engine Fuel Type'].isna())  & (df_train['Make'] == 'Suzuki'), 'Engine Fuel Type'] = 'Regular unleaded'
```

```python
df_train.isna().sum()
```

```python
df_test.isna().sum()
```

```python
# fill missing values of each column with the mode of the column
df_train['Engine HP'].fillna(df_train['Engine HP'].mode()[0], inplace = True)
df_train['Engine Cylinders'].fillna(df_train['Engine Cylinders'].mode()[0], inplace = True)
df_train['Market Category'].fillna(df_train['Market Category'].mode()[0], inplace = True)
df_test['Engine HP'].fillna(df_train['Engine HP'].mode()[0], inplace = True)
df_test['Engine Cylinders'].fillna(df_train['Engine Cylinders'].mode()[0], inplace = True)
df_test['Market Category'].fillna(df_train['Market Category'].mode()[0], inplace = True)
```

```python
df_train.isna().sum()
```

```python
df_test.isna().sum()
```

```python
df_train.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder

encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
for col in ['Make', 'Model', 'Engine Fuel Type', 'Transmission Type', 'Driven_Wheels', 'Market Category', 'Vehicle Size', 'Vehicle Style']:
    df_train[col] = encoder.fit_transform(df_train[[col]])
    df_test[col] = encoder.transform(df_test[[col]])
```

```python
df_train.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

for col in df_train.columns:
    if col != 'MSRP':
        df_train[col] = scaler.fit_transform(df_train[[col]])
        df_test[col] = scaler.transform(df_test[[col]])
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
X = df_train.drop('MSRP', axis = 1)
y = df_train['MSRP']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=.1, random_state=313)
```

<h2 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
مدل‌سازی
</font>
</h2>

<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
<font face="vazir" size=3>
    حال که داده را پاکسازی کرده و احتمالا ویژگی‌هایی را به آن افزوده یا از آن حذف کرده‌اید، وقت آن است که مدلی آموزش دهید که بتواند متغیر هدف این مسئله را پیش‌بینی کند.
</font>
</p>

<h3 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
آموزش مدل
</font>
</h3>

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
    در سلول پایین با استفاده از کتابخانه <code>scikit-learn</code> مدلی آموزش دهید که بتواند متغیر هدف این مسئله را پیش‌بینی کند.
</font>
</p>

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
pol = PolynomialFeatures(degree=2, include_bias=False, interaction_only=False)
model = LinearRegression()
model.fit(pol.fit_transform(X_train), y_train)
```

<h3 align=right style="line-height:200%;font-family:vazir;color:#0099cc">
<font face="vazir" color="#0099cc">
معیار ارزیابی
</font>
</h3>

<p dir=rtl style="direction: rtl; text-align: justify; line-height:200%; font-family:vazir; font-size:medium">
<font face="vazir" size=3>
    معیاری که برای ارزیابی عملکرد مدل انتخاب کرده‌ایم، <code>r2_score</code> نام دارد.
    <br>
    پیشنهاد می‌شود با توجه به این معیار، عملکرد مدل خود را بر روی مجموعه‌ی آموزش یا اعتبارسنجی ارزیابی کنید و طبق نتایج به‌دست‌آمده پارامترهای مدل خود را بهتر تنظیم کنید.
</font>
</p>

```python
# evaluate model
from sklearn.metrics import r2_score

y_pred = model.predict(pol.transform(X_val))
r2_score(y_val, y_pred)
```

```python
y_pred = model.predict(pol.transform(X_train))
r2_score(y_train, y_pred)
```

```python
from flaml import AutoML

automl = AutoML(task='regression', time_budget=60)
automl.fit(pol.fit_transform(X_train), y_train)
```

```python
y_pred = automl.predict(pol.transform(X_train)) 
r2_score(y_train, y_pred)
```

```python
y_pred = automl.predict(pol.transform(X_val)) 
r2_score(y_val, y_pred)
```
